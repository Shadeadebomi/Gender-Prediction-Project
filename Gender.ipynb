# Gender Prediction Using Machine Learning (Classification)

## 1. Project Overview

This project demonstrates a **classification machine learning model** used to predict **gender (Male or Female)** based on facial features.

Because the target variable (**gender**) has two categories (Male / Female), this is a **Classification problem**, not regression.

---

## 2. Understanding the Dataset (Beginner Explanation)

Each row represents one person with facial characteristics:

* **long_hair**: 1 = Yes, 0 = No
* **forehead_width_cm**: Width of forehead in centimeters
* **forehead_height_cm**: Height of forehead in centimeters
* **nose_wide**: 1 = Wide nose, 0 = Narrow
* **nose_long**: 1 = Long nose, 0 = Short
* **lips_thin**: 1 = Thin lips, 0 = Thick lips
* **distance_nose_to_lip_long**: 1 = Long distance, 0 = Short
* **gender**: Target variable (Male or Female)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
```

```python
## 8. Feature Encoding

#Machine learning models only understand numbers.

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])  # Male=1, Female=0

```

```python

## 9. Feature Scaling
X = df.drop('gender', axis=1)
y = df['gender']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## 10. Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

## 11. Model Training

```python
### 11.1 Logistic Regression

log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_pred = log_model.predict(X_test)
```

```python

### 11.2 K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
knn_pred = knn_model.predict(X_test)
```

```python
### 11.3 Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
```

```python

### 11.4 Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
## 12. Model Evaluation

results = [
    {'Model': 'Logistic Regression', 'Accuracy': accuracy_score(y_test, log_pred)},
    {'Model': 'KNN', 'Accuracy': accuracy_score(y_test, knn_pred)},
    {'Model': 'Decision Tree', 'Accuracy': accuracy_score(y_test, dt_pred)},
    {'Model': 'Random Forest', 'Accuracy': accuracy_score(y_test, rf_pred)}
]

results_df = pd.DataFrame(results)
results_df.sort_values(by='Accuracy', ascending=False)


```

```python
## 13. Confusion Matrix (Best Model)
sns.heatmap(confusion_matrix(y_test, rf_pred), annot=True, fmt='d')

```

```python
## 14. Testing with New Input Data

new_person = pd.DataFrame({
    'long_hair': [1],
    'forehead_width_cm': [13.8],
    'forehead_height_cm': [6.0],
    'nose_wide': [0],
    'nose_long': [0],
    'lips_thin': [0],
    'distance_nose_to_lip_long': [0]
})

new_person_scaled = scaler.transform(new_person)
prediction = rf_model.predict(new_person_scaled)

print('Male' if prediction[0] == 1 else 'Female')
```
